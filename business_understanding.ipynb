# Business Understanding and Analytic Approach

## 1. Problem Introduction

The rapid growth of e-commerce has significantly increased the number of orders, while also creating a major challenge in product returns and reverse logistics. In this context, reverse logistics is not only an operational cost issue but also affects customer experience, supply chain efficiency, and circular economy goals.

This project focuses on analyzing e-commerce order data to identify which factors increase return risk. Instead of treating reverse logistics as a purely operational process, this analysis views it as a data problem that can be predicted and explained using features such as delivery time, shipping cost, product category, payment method, and customer reviews.

## 2. Analytical Objectives

This project aims to answer the following research questions:

- Which factors are most strongly associated with product return risk in e-commerce?
- Does late delivery increase the likelihood of returns?
- Do shipping cost, order value, product type, and payment method affect return behavior?
- Can customer review scores reflect return risk?
- Can a logistic regression model be built to predict the probability of returns from historical data?

The final goal is to produce an analysis that is both practically useful for reverse logistics and concise enough to be presented in a 9–12 page paper.

## 3. Scope and Unit of Analysis

The dataset used in this project is the Brazilian E-Commerce Public Dataset by Olist, a public dataset containing around 100,000 orders from 2016–2018, including tables related to orders, products, payments, customers, sellers, reviews, and geolocation.

The main unit of analysis in this project is one order. Each order will be enriched with relevant attributes from other tables to create a final analytical dataset for SQL, EDA, and regression analysis. This approach preserves the original relational structure while still allowing a clean master table for modeling.

## 4. Potential Influencing Factors

Based on the Olist dataset structure and the reverse logistics context, the following factors may influence return risk:

- Delivery time: late deliveries may increase customer dissatisfaction.
- Shipping cost: higher shipping cost may make customers more sensitive to the overall service experience.
- Order value: high-value orders may show different return behavior from low-value ones.
- Product category: some categories, such as fashion, accessories, or small household items, may have higher return rates.
- Payment method: installment payments, credit card payments, or cash payments may be associated with different return patterns.
- Review score: low ratings may reflect dissatisfaction and may correlate with return behavior.
- Customer and seller location: geographic distance or logistical complexity may affect delivery performance and returns.

These factors support not only prediction but also operational interpretation, helping connect the analysis to the broader goal of reducing waste and optimizing reverse logistics in a circular economy.

## 5. Analytical Approach

The project will follow a six-step workflow, matching the structure required by the instructor:

1. **Business Understanding and Analytic Approach**
   - Define the business problem, research scope, and research questions.
   - Choose the target variable and define how “return risk” will be measured.

2. **Data Collection, Understanding, Preparation**
   - Read the Olist CSV files.
   - Check table structures, join keys, missing values, and duplicate values.
   - Build the final analytical table from the relevant source tables.

3. **Data Analysis with SQL**
   - Use SQL to join data and explore order distribution, product groups, delivery status, and reviews.

4. **Data Analysis with Python**
   - Use pandas and NumPy for cleaning, transformation, and initial analysis.
   - Create derived variables such as delay days, shipping duration, review groups, and product categories.

5. **Data Visualization**
   - Use charts to examine the relationship between return risk and explanatory variables.
   - Compare return risk across categories, delivery times, review scores, and payment types.

6. **Regression Analysis**
   - Build a logistic regression model to predict the probability of returns.
   - Interpret model coefficients to identify the most influential factors affecting return risk.

## 6. Expected Outcomes

The project is expected to achieve the following outcomes:

- Identify the factors most strongly associated with product return risk.
- Understand the roles of delivery time, shipping cost, product category, and review score in return behavior.
- Build a logistic regression baseline that is clear and suitable for inclusion in the paper.
- Provide practical implications for reverse logistics, such as reducing late deliveries, improving handling of high-risk product categories, or prioritizing orders with higher return probability.


## 7. Practical Significance

The results of this project can be useful for stakeholders in e-commerce. For businesses, the model may help identify orders with a high risk of return so that operational actions and customer support strategies can be adjusted in advance. From an academic perspective, the project shows how e-commerce transaction data can be used to study reverse logistics in the context of the circular economy.
